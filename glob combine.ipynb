```python
#Use glob to match the pattern ‘csv’
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#Combine all files in the list and export as CSV
jan = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
jan.to_csv( "1/combined_csv.csv", index=False, encoding='utf-8-sig')
```

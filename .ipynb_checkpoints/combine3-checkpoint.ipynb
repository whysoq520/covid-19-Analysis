```python
#Use glob to match the pattern ‘csv’
#extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format('csv'))]
#Combine all files in the list
data1= pd.concat([pd.read_csv(f) for f in all_filenames ])
data1.head(2)
```

```python
data2 = data1.drop(['Active', 'Admin2','Combined_Key', 'FIPS'], axis = 1)
data2.columns
```

```python
data2.columns = ['Province_State', 'Country_Region', 'Last_Update', 'Latitude', 'Longitude',
       'Confirmed', 'Deaths', 'Recovered']
data2.head(2)
```

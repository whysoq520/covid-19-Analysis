```python
import pandas as pd
```

```python
df_country = pd.read_csv('./Resources/Data/country/country.csv')
df_country
```

```python
df_country.shape
```

```python
df_country = df_country.copy()
for index, row in df_country.iterrows():
    try:
        # Target city
        target_country = row['Name']
        
        # Build the endpoint URL
        url = f"https://maps.googleapis.com/maps/api/geocode/json?key={gkey}&address={target_country}"

        # Run a request to endpoint and convert result to json
        geo_data = requests.get(url).json()

        # Extract latitude and longitude
        lat = geo_data['results'][0]['geometry']['location']['lat']
        lng = geo_data['results'][0]['geometry']['location']['lng']
        
        df_country.loc[df_country.Name == target_country, ['Lat', 'Lon']] = lat, lng
        
    except:
        print(target_country)
        pass
```
